```python
def correlation_check(traindata,testdata,thresh): # drop columns above certain threshold    
        corr_matrix = traindata.corr()
        upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(np.bool))
        to_drop = [column for column in upper.columns if any(upper[column] >thresh)]
        trainset=traindata.drop(traindata[to_drop], axis=1)
        testset=testdata.drop(testdata[to_drop],axis=1)
        return trainset,testset
```

```python
train_corr , test_corr = correlation_check(combined_features_train_sweet, combined_features_test_sweet,0.7)
```

```python
y_train_corr = train_corr['Sweet']
y_train_corr
```

```python
x_train_corr = train_corr.drop(['Sweet','Name','Name.1','Taste'],axis=1)
x_train_corr
```

```python
y_test_corr = test_corr['Sweet']
y_test_corr
```

```python
x_test_corr = test_corr.drop(['Sweet','Name','Name.1','Taste'],axis=1)
x_test_corr
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(x_train_corr)
new_x_train_corr_without_tar = scaler.transform(x_train_corr)
```

```python
scaler = StandardScaler()
scaler.fit(x_test_corr)
new_x_test_corr_without_tar = scaler.transform(x_test_corr)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc2=RandomForestClassifier(random_state=42, max_features='auto', n_estimators= 200, max_depth=8, criterion='entropy')
```

```python
rfc2.fit(new_x_train_corr_without_tar,y_train_corr)
```

```python
y_pred_corr_train_without_tar = rfc2.predict(new_x_train_corr_without_tar)
y_pred_corr_train_without_tar
```
